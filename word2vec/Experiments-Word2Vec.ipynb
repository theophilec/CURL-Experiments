```python
from gensim.models import Word2Vec 
import gensim.downloader as api
from sklearn.manifold import TSNE
import re
import matplotlib.pyplot as plt
import numpy as np

v2w_model = api.load('word2vec-google-news-300')
```

```python
model = v2w_model
```

```python
X = model[list(model.wv.vocab)]
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X[:1000,:])
```

```python
l = list(model.wv.vocab)
```

```python
months_list = ["October", "September","April"]
gender = [ "father", "mother", "man", "woman"]
work = ["work", "job"]
food = ["cheese"]
countries = ["Iran", "Pakistan", "England", "Europe"]
politics = ["House", "Senate", "Obama", "federal"]
```

```python
#Note: tSNE is stochastic, and we have not fixed its seed, to figures may not exactly reproduce.
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], color="grey", alpha=0.3)

for word in months_list:
    plt.scatter(
        X_tsne[l.index(word), 0], X_tsne[l.index(word), 1], label=word, color="blue"
    )
for word in gender:
    plt.scatter(
        X_tsne[l.index(word), 0], X_tsne[l.index(word), 1], label=word, color="red"
    )
for word in work:
    plt.scatter(
        X_tsne[l.index(word), 0], X_tsne[l.index(word), 1], label=word, color="green"
    )
for word in countries:
    plt.scatter(
        X_tsne[l.index(word), 0], X_tsne[l.index(word), 1], label=word, color="black"
    )
for word in politics:
    plt.scatter(
        X_tsne[l.index(word), 0], X_tsne[l.index(word), 1], label=word, color="orange"
    )
# plt.legend()
plt.savefig("projection.png", dpi=500)
```
